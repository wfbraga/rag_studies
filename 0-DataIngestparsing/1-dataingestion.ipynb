## Understandi Document Structure in LangChain

```python
# Creating a simple document for testing
doc = Document(
    page_content="This is a sample document for testing the text splitters in LangChain.",
    metadata={
        "source": "test_document",
        "author": "ChatGPT",
        "page": 1,
        "date_created": "2026-06-01"
        }
)
print(f"Document created: {doc.page_content[:50]}... with metadata: {doc.metadata}")
why_metadata="""
Metadata é crucial para:
  1 Filtrar busca,
  2 Rastrear as fontes dos documentos,
  3 Dar contexto adicional para o conteúdo do documento,
  4 "Debugging" e auditoria de dados.
"""
print(why_metadata)
```

```python
print(f"Agora doc é um {type(doc)}")
print(f"doc.page_content é um {type(doc.page_content)}")
print(f"doc.metadata é um {type(doc.metadata)}")
```
